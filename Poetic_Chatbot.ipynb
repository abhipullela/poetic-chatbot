```python
import google.generativeai as genai
import time
from google.colab import userdata

#importing the required libraries for our chatbot to work
```

```python
key = userdata.get("key") #obtaining the gemini api key from the secrets tab in google collab
```

```python
genai.configure(api_key=key) #configuration process
```

```python
model = genai.GenerativeModel("gemini-2.0-flash") #selection of the required model for the generation of content
```

```python
def text_flow(text):
    for word in text.split():
        print(word, end=" ", flush=True)
        time.sleep(0.15)
    print()

#the fancy typing function for smooth text flow
```

```python
def poetic_response(user_text):
    prompt = f"""
Transform this message into poetic, emotional verse (give me only the verse - short and crisp):

User: {user_text}

Poem:
    """

    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Error: {e}"

#giving the required prompt to the gemini version chosen via formattted printing
#try-except used to gracefully tackle any kinds of errors
```

```python
print("Welcome to your Poetic Guide!")
print("Type 'quit' to exit")
while True:
    user = input("You: ")
    if user.lower() in ["exit", "quit"]:
        print("Chatbot: May your thoughts wander in gentle verses :)")
        break

    poem = poetic_response(user)
    print("Chatbot:", end=" ")
    text_flow(poem)

#the final main menu of our project
```
